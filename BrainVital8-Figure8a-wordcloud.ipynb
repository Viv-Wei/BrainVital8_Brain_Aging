```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Category wordcloud — Nature 
pip install pandas wordcloud matplotlib numpy pillow
"""

import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np

# =====================================================================
# 1. reading data
# =====================================================================
data = pd.read_csv(
    r"./data/BrainVital8/SNP_5e_8_unique_outof_ENSG_0312_classified.csv",
    low_memory=False,
)
category_counts = data["Category_merged"].value_counts()
category_dict = category_counts.to_dict()

# =====================================================================
# 2. Nature color scheme
# =====================================================================
highlight_categories = {
    "Body and muscle composition":     "#C44E52",  
    "Respiratory function":            "#4C72B0",  
    "Brain structure and function":    "#17807E",  
    "Immune cells":                    "#55A868",  
    "Celluar Energy metabolism":       "#DD8452",  
    "Cellular energy metabolism":      "#DD8452",  
    "Glucose metabolism and diabetes": "#E5A02E",  
    "Aging":                           "#8B5E3C",  
    "Behaviour factors":               "#D4527A",  
    "Depression":                      "#6A9E3A",  
    "Sleep quality":                   "#8172B3",  
    "Smoking":                         "#CCB974",  
    "Social factors":                  "#3A8FB7",  
    "Nervous System Homeostasis":      "#A65D8C",  
    "Alzheimer disease":               "#D45B3E",  
}

BACKGROUND_COLOR = "#C8C8C8"

color_map = {}
for cat in category_dict.keys():
    color_map[cat] = highlight_categories.get(cat, BACKGROUND_COLOR)

def nature_color_func(word, **kwargs):
    return color_map.get(word, "#8C8C8C")

# =====================================================================
# 3. wordcloud generation
# =====================================================================
mask_w, mask_h = 2400, 2600  
mask = np.zeros((mask_h, mask_w), dtype=np.uint8)  

wordcloud = WordCloud(
    width=mask_w,
    height=mask_h,
    font_path="C:/Windows/Fonts/Arial.ttf",  # Mac: "/Library/Fonts/Arial.ttf"
    background_color="white",
    max_words=800,
    prefer_horizontal=1.0,
    relative_scaling=0.15,
    min_font_size=1,
    max_font_size=160,
    random_state=42,
    collocations=False,
    colormap=None,
    margin=0,
    mode="RGB",
    repeat=False,
    mask=mask,
).generate_from_frequencies(category_dict)

wordcloud_recolored = wordcloud.recolor(color_func=nature_color_func)

# =====================================================================
# 4. exporting
# =====================================================================
# SVG   
svg_content = wordcloud_recolored.to_svg(embed_font=True)
with open("category_wordcloud_nature.svg", "w", encoding="utf-8") as f:
    f.write(svg_content)
print("save as: category_wordcloud_nature.svg")

# PDF（pip install cairosvg）
try:
    import cairosvg
    cairosvg.svg2pdf(
        bytestring=svg_content.encode("utf-8"),
        write_to="category_wordcloud_nature.pdf",
    )
    print("save as: category_wordcloud_nature.pdf")
except ImportError:
    print("please: pip install cairosvg")

# PNG 
fig, ax = plt.subplots(figsize=(12, 13))
ax.imshow(wordcloud_recolored, interpolation="bilinear")
ax.axis("off")
plt.tight_layout()
plt.savefig("category_wordcloud_nature.png", format="png", bbox_inches="tight", dpi=300)
plt.show()
print("save as: category_wordcloud_nature.png")
```
